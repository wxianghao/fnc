## Example 2.6.1


```julia
using FNCFunctions
```

```julia
A = [2 0 4 3 ; -4 5 -7 -10 ; 1 15 2 -4.5 ; -2 0 2 -13];
L, U = FNC.lufact(A)
L
```

```julia
A[[2, 4], :] = A[[4, 2], :]  
L, U = FNC.lufact(A)
L
```

```julia
U[1, :] = A[1, :]
L[:, 1] = A[:, 1] / U[1, 1]
A -= L[:, 1] * U[1, :]'
```


## Example 2.6.2

```julia
A₁ = [2 0 4 3 ; -2 0 2 -13; 1 15 2 -4.5 ; -4 5 -7 -10]
```

```julia
L, U = zeros(4,4), zeros(4,4)
```

```julia
i = argmax( abs.(A₁[:, 1]) ) 
```

```julia
U[1, :] = A₁[i, :]
L[:, 1] = A₁[:, 1] / U[1, 1]
A₂ = A₁ - L[:, 1] * U[1, :]'
```

```julia
i = argmax( abs.(A₂[:, 2]) ) 
```

```julia
U[2, :] = A₂[i, :]
L[:, 2] = A₂[:, 2] / U[2, 2]
A₃ = A₂ - L[:, 2] * U[2, :]'
```

```julia
@show i = argmax( abs.(A₃[:, 3]) ) 
U[3, :] = A₃[i, :]
L[:, 3] = A₃[:, 3] / U[3, 3]
A₄ = A₃ - L[:, 3] * U[3, :]'
```

```julia
@show i = argmax( abs.(A₄[:, 4]) ) 
U[4, :] = A₄[i, :]
L[:, 4] = A₄[:, 4] / U[4, 4];
```

```julia
A₁ - L * U
```

## Example 2.6.4

```julia
using FNCFunctions
A = rand(1:20, 4, 4)
L, U, p = FNC.plufact(A)
A[p, :] - L * U   # should be ≈ 0
```

```julia
b = rand(4)
z = FNC.forwardsub(L, b[p])
x = FNC.backsub(U, z)
```

```julia
b - A*x
```

## Example 2.6.5

## Example 2.6.6

```julia
ϵ = 1e-12
A = [-ϵ 1; 1 -1]
b = A * [1, 1]
```

```julia
using FNCFunctions
L, U = FNC.lufact(A)
x = FNC.backsub( U, FNC.forwardsub(L, b) )
```

```julia
L, U, p = FNC.plufact(A)
x = FNC.backsub( U, FNC.forwardsub(L, b[p]) )
```
