## Example 1.4.2

```julia
a, b, c = 1., -(1e6 + 1e-6), 1.
disc = sqrt(b^2 - 4 * a * c)
x₁ = (-b + disc) / 2a
```

```julia
abs(x₁ - 1e6) / 1e6
```

Use $x_1 x_2 = \dfrac{c}{a}$

```julia
x₂ = c / (a * x₁)
```

```julia
rel_error = abs(x₂ - 1e-6) / 1e-6
```

## Example 1.4.3

```julia
using Polynomials
r = [-2.0, -1, 1, 1, 3, 6]
p = fromroots(r)
```

```julia
r̃ = sort(roots(p))   # type r\tilde and then press Tab
```

Relative errors in the roots:

```julia
@. abs(r - r̃) / r
```

```julia
rel_error = @. abs(r - r̃) / r
```

The polynomial whose roots we actually got:

```julia
p̃ = fromroots(r̃)
```

Backward error:

```julia
coeffs(p) - coeffs(p̃)
```
