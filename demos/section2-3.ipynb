## Example 2.3.3


### forward substitution (lower triangular)

```julia
A = rand(1.:9., 5, 5)
```

```julia
using LinearAlgebra
L = tril(A)
```

```julia
b = ones(5)
```

solve $\mathbf{L} \mathbf{x} = \mathbf{b}$

```julia
using FNCFunctions
x = FNC.forwardsub(L, b)
```

residual:

```julia
b - L * x
```

### backward substitution (upper triangular)

```julia
α = 0.3
β = 2.2
U = diagm( 0=>ones(5), 1=>[-1, -1, -1, -1] )
U[1, [4, 5]] = [ α - β, β ]
U
```

```julia
x_exact = ones(5)
b = [α, 0, 0, 0, 1]
```

```julia
x = FNC.backsub(U,b)
err = x - x_exact
```

```julia
α = 0.3;
β = 1e12;
U = diagm( 0=>ones(5), 1=>[-1, -1, -1, -1] )
U[1, [4, 5]] = [ α - β, β ]
```

```julia
x = FNC.backsub(U, b)
err = x - x_exact
```
