## Example 1.2.5

Quadratic polynomial $p(x) = \frac{1}{3}(x-1)(x-1-\epsilon)$

```julia
ε = 1e-6
a = 1/3
b = (-2 - ε) / 3
c = (1 + ε) / 3
```

```julia
a, b, c
```

```julia
disc = sqrt(b^2 - 4 * a * c)
x₁ = (-b + disc) / 2a
x₂ = (-b - disc) / 2a
x₁, x₂
```

### Relative errors in the roots

```julia
r₁, r₂ = 1 + ε, 1
```

```julia
abs(r₁ - x₁) / abs(r₁)
```

```julia
abs(r₂ - x₂) / abs(r₂)
```

### Condition numbers

```julia
κ₁ = abs(r₁) / abs(r₁ - r₂)
```

```julia
κ₁ * eps()
```

```julia
κ₂ = abs(r₂) / abs(r₁ - r₂)
```

```julia
κ₂ * eps()
```
