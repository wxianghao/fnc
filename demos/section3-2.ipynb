## Example 3.2.1

$$
\sin^2(t) + \cos^2(t) - 1 = 0
$$

```julia
t = range(0, 3, 400)
f = [ x -> sin(x)^2, x -> cos((1 + 1e-7) * x)^2, x -> 1. ]
A = [ f(t) for t in t, f in f ]
```

```julia
using LinearAlgebra
κ = cond(A)
```

```julia
x = [1., 2, 1]
b = A * x;
```

```julia
x_BS = A \ b
```

```julia
normres = norm(b - A * x_BS) / norm(b)
```

```julia
observed_error = norm(x_BS - x) / norm(x)
```

```julia
error_bound = κ * eps()
```

```julia
N = A' * A
x_NE = N \ (A'*b)
```

```julia
observed_err = norm(x_NE - x) / norm(x)
```

```julia
digits = -log10(observed_err)
```
