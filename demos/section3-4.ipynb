## Example 3.4.1

```julia
A = rand(float(1:9), 6, 4)
m,n = size(A)
```

```julia
using LinearAlgebra
z = A[:, 1];
v = normalize(z - norm(z) * [1; zeros(m-1)])
P₁ = I - 2v * v'   # reflector
```

New structural zeros:

```julia
P₁ * z
```

```julia
A = P₁ * A
```

```julia
z = A[2:m, 2]
v = normalize(z - norm(z) * [1; zeros(m-2)])
P₂ = I - 2v * v'
```

```julia
A[2:m, :] = P₂ * A[2:m, :]
A
```

```julia
for j in 3:n
    z = A[j:m, j]
    v = normalize(z - norm(z) * [1; zeros(m-j)])
    P = I - 2v * v'
    A[j:m, :] = P * A[j:m, :]
end
```

```julia
R = triu(A)
```
