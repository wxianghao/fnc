## Example 2.4.3

```julia
using LinearAlgebra

A₁ = [
     2    0    4     3 
    -4    5   -7   -10 
     1   15    2   -4.5
    -2    0    2   -13
    ];
```

```julia
L = diagm(ones(4))
```

```julia
U = zeros(4, 4)
```

$\mathbf{A}_1 = \boldsymbol{\ell}_1\mathbf{u}_1^T + \boldsymbol{\ell}_2\mathbf{u}_2^T + \boldsymbol{\ell}_3\mathbf{u}_3^T + \boldsymbol{\ell}_4\mathbf{u}_4^T$

```julia
U[1, :] = A₁[1, :]
U
```

```julia
L[:, 1] = A₁[:, 1] / U[1, 1]
L
```

```julia
A₂ = A₁ - L[:, 1] * U[1, :]'
```

$\mathbf{A}_2 = \boldsymbol{\ell}_2\mathbf{u}_2^T + \boldsymbol{\ell}_3\mathbf{u}_3^T + \boldsymbol{\ell}_4\mathbf{u}_4^T$ 

```julia
U[2, :] = A₂[2, :]
U
```

```julia
L[:, 2] = A₂[:, 2] / U[2, 2]
L
```

```julia
A₃ = A₂ - L[:, 2] * U[2, :]'
```

Now we can deal with the lower right $2\times 2$ submatrix of the remainder in a similar fashion.

```julia
U[3, :] = A₃[3, :]
L[:, 3] = A₃[:, 3] / U[3, 3]
A₄ = A₃ - L[:, 3] * U[3, :]'
```

Finally, we pick up the last unknown in the factors.

```julia
U[4, 4] = A₄[4, 4];
```

```julia
L
```

```julia
U
```

Backward error:

```julia
A₁ - L * U
```

## Example 2.4.4

```julia
A = [2 0 4 3; -4 5 -7 -10; 1 15 2 -4.5; -2 0 2 -13];
b = [4, 9, 9, 4];
```

```julia
A
```

```julia
b
```

```julia
using FNCFunctions
L, U = FNC.lufact(A);
```

$$
\begin{aligned}
\mathbf{A} \mathbf{x} &= \mathbf{b} \\ 
(\mathbf{L} \mathbf{U}) \mathbf{x} &= \mathbf{b}\\
\mathbf{L} (\mathbf{U} \mathbf{x}) &= \mathbf{b} \\
\mathbf{U} \mathbf{x} = \mathbf{z} \quad & \mathrm{ and } \quad \mathbf{L} \mathbf{z} = \mathbf{b}
\end{aligned}
$$


```julia
z = FNC.forwardsub(L, b)
x = FNC.backsub(U, z);
```

```julia
b - A*x
```
