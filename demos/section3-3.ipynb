## Example 3.3.2

```julia
t = range(0, 3, 400)
f = [ x -> sin(x)^2, x -> cos((1 + 1e-7) * x)^2, x -> 1. ]
A = [ f(t) for t in t, f in f ];
```

```julia
using LinearAlgebra
κ = cond(A)
```

```julia
x = [1., 2, 1]
b = A * x;
```

```julia
using FNCFunctions
x_QR = FNC.lsqrfact(A, b)
observed_error = norm(x_QR - x) / norm(x)
```

```julia
error_bound = κ * eps()
```
