## Example 2.9.2

```julia
A₁ = [  2     4     4     2.0
        4     5     8    -5
        4     8     6     2
        2    -5     2   -26   ]
```

```julia
using LinearAlgebra
L = LowerTriangular(diagm(ones(4)))
```

```julia
d = zeros(4)
```

```julia
d[1] = A₁[1, 1]
L[:, 1] = A₁[:, 1] / d[1]
A₂ = A₁ - d[1] * L[:, 1] * L[:, 1]'
```

```julia
d[2] = A₂[2, 2]
L[:, 2] = A₂[:, 2] / d[2]
A₃ = A₂ - d[2] * L[:, 2] * L[:, 2]'
```

```julia
d[3] = A₃[3, 3]
L[:, 3] = A₃[:, 3] / d[3]
A₄ = A₃ - d[3] * L[:, 3] * L[:, 3]'
```

```julia
d[4] = A₄[4, 4]
@show d;
L
```

```julia
opnorm(A₁ - (L * diagm(d) * L'))
```

## Example 2.9.3

```julia
A = rand(1.0:9.0, 4, 4)
B = A + A'
```

$$(A + A^T)^T = A^T + (A^T)^T = A^T + A$$

```julia
using LinearAlgebra
cholesky(B)    # throws an error
```

```julia
B = A' * A
cf = cholesky(B)
```

```julia
R = cf.U
```

```julia
opnorm(R' * R - B) / opnorm(B)
```
