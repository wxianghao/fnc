## Example 2.1.1

Population of China

```julia
year = [1982, 2000, 2010, 2015]; 
pop = [1008.18, 1262.64, 1337.82, 1374.62];
```

time = years since 1980

```julia
t = year .- 1980.0
y = pop;
```

$n=4 \implies$ cubic polynomial

```julia
V = [ t[i]^j for i in 1:4, j in 0:3 ]
```

Solve $Vc=y$ for $c$

```julia
c = V \ y
```

Residual (close to zero?)

```julia
y - V * c
```

```julia
using Polynomials
p = Polynomial(c)   # construct a polynomial
```

Official pop in 2005 = 1,303,720,000

```julia
p(2005-1980)    # include the 1980 time shift
```

```julia
using Plots
theme(:dark)
```

```julia
scatter(t, y;
    label="actual",  legend=:topleft,
    xlabel="years since 1980",  
    ylabel="population (millions)", 
    title="Population of China")
```

```julia
tt = range(0, 35, 500)   # 500 times in [0, 35]
yy = p.(tt)              # evaluate p at all
```

```julia
size(yy)
```

```julia
yy[1:4]
```

```julia
plot!(tt, yy, label="interpolant")
```
