## Example 4.5.3


$$
\begin{split}
    f_1(x_1,x_2,x_3) &= e^{x_2 - x_1} - 2 \\
    f_2(x_1,x_2,x_3) &= x_1x_2 + x_3\\
    f_3(x_1,x_2,x_3) &= x_2x_3 + x_1^2 - x_2
\end{split}
$$

$$
    \mathbf{J}(x) =
    \begin{bmatrix}
       -e^{x_2 - x_1} & e^{x_2 - x_1} & 0\\
      x_2 & x_1 & 1\\
       2x_1 & x_3-1 & x_2
    \end{bmatrix}.
$$


```julia
function func(x)
    return [exp(x[2] - x[1]) - 2,
        x[1] * x[2] + x[3],
        x[2] * x[3] + x[1]^2 - x[2]
    ]
end;

function jac(x)
    J = zeros(3, 3)
    J[:, 1] = [-exp(x[2] - x[1]), x[2], 2x[1]]
    J[:, 2] = [ exp(x[2] - x[1]), x[1], x[3] - 1]
    J[:, 3] = [0, 1, x[2]]
    return J
end;
```

```julia
x = [-1, 2, 3]
δ = 1e-5
(func(x + δ*[0, 0, 1]) - func(x)) / δ
```

```julia
jac(x)
```

```julia
using FNCFunctions

x₁ = [0, 0, 0]
ϵ = eps()
x = FNC.newtonsys(func, jac, x₁, xtol=ϵ, ftol=ϵ);
```

```julia
x
```

```julia
func(x[end])
```

```julia
using LinearAlgebra

x₁ = BigFloat.([0, 0, 0])
ϵ = eps(BigFloat)
x = FNC.newtonsys(func, jac, x₁, xtol=ϵ, ftol=ϵ);
r = x[end]
residual = norm(func(r))
```

```julia
using PrettyTables

logerr = [Float64(log(norm(r - x[k]))) for k in 1:length(x)-1]
ratios = [NaN; [logerr[i+1] / logerr[i] for i in 1:length(logerr)-1]]

pretty_table( (iter=eachindex(logerr), logerr, ratios);
    column_labels=["iteration", "log error", "ratio"], backend=:html )
```
