## Example 2.7.3

```julia
A = [ 2 0; 1 -1 ]
```

```julia
using LinearAlgebra
norm(A)
```

```julia
opnorm(A)
```

```julia
opnorm(A, 1)
```

```julia
# Sum down the rows (1st matrix dimension):
maximum( sum(abs.(A), dims=1) ) 
```

```julia
opnorm(A, Inf)
```

```julia
 # Sum across columns (2nd matrix dimension):
maximum( sum(abs.(A), dims=2) ) 
```

```julia
# Construct 601 unit column vectors.
θ = 2π * (0:1/600:1)   # type \theta then Tab
x = [ fun(t) for fun in [cos, sin], t in θ ];
size(x)
```

```julia
using Plots, LaTeXStrings
plot(aspect_ratio=1, layout=(1, 2), xlabel=L"x_1",  ylabel=L"x_2", legend=false)
plot!(x[1, :], x[2, :], subplot=1, title="Unit circle") 
```

```julia
Ax = A * x;
size(Ax)
```

```julia
plot!(Ax[1, :], Ax[2, :], subplot=2, title="Image under x → Ax")
```

```julia
r = opnorm(A);
plot!(r * cos.(θ), r * sin.(θ), subplot=2, l=:dash)
```
