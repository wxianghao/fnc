## Example 1.1.3

Integers and floating-point values are different.

```julia
typeof(1)
```

```julia
typeof(1.0)
```

To go from float to integer, use `Int`.

```julia
Int(1.0)
```

That's an error if the conversion would change the value.

```julia
Int(1.0001)
```

Internal binary representation of floats:

```julia
bitstring(1.0)
```

```julia
bitstring(-1.0)
```

```julia
bitstring(2.0)
```

```julia
bitstring(0.5)
```

```julia
bitstring(1 + eps())
```

Getting the 3 parts of a float (as base-10 values):

```julia
A = 6.022e23
sign(A), significand(A), exponent(A)
```

```julia
significand(A) * 2.0 ^ exponent(A)
```

```julia
1.0 + eps() / 2
```
