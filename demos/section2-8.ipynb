## Example 2.8.1

```julia
[ 1 // (i + j) for i in 1:6, j in 1:6 ]
```

```julia
A = [ 1 / (i + j) for i in 1:6, j in 1:6 ]
```

```julia
using LinearAlgebra
κ = cond(A)
```

```julia
x = 1:6
b = A * x
```

```julia
using Random
Random.seed!(302)
```

```julia
# type \Delta then Tab to get Δ
ΔA = randn(size(A));  ΔA = 1e-12 * (ΔA / opnorm(ΔA));
Δb = randn(size(b));  Δb = 1e-12 * (Δb / norm(Δb));
```

```julia
new_x = ((A + ΔA) \ (b + Δb))
Δx = new_x - x
```

```julia
@show relative_error = norm(Δx) / norm(x);
```

Upper bound due to perturbations of $b$:

```julia
κ * norm(Δb) / norm(b)
```

Upper bound due to perturbations of $A$:

```julia
κ * opnorm(ΔA) / opnorm(A)
```

Perturbations due to machine roundoff:

```julia
Δx = A\b - x
@show relative_error = norm(Δx) / norm(x);
@show rounding_bound = κ * eps();
```
